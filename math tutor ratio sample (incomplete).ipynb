```python
import pandas as pd
import numpy as np
import csv as csv
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('algebra_2005_2006_train.txt', sep='\t')
```

```python
df[['KC1', 'KC2', 'KC3', 'KC4', 'KC5', 'KC6', 'KC7']] = df['KC(Default)'].str.split('~~', expand=True)
```

```python
df[['Opps1', 'Opps2', 'Opps3', 'Opps4', 'Opps5', 'Opps6', 'Opps7']] = df['Opportunity(Default)'].str.split('~~', expand=True)
```

```python
df['Opps1'] = pd.to_numeric(df['Opps1'])
df['Opps2'] = pd.to_numeric(df['Opps2'])
df['Opps3'] = pd.to_numeric(df['Opps3'])
df['Opps4'] = pd.to_numeric(df['Opps4'])
df['Opps5'] = pd.to_numeric(df['Opps5'])
df['Opps6'] = pd.to_numeric(df['Opps6'])
df['Opps7'] = pd.to_numeric(df['Opps7'])
```

```python
df = df.drop('Step Start Time', 1)
df = df.drop('Step End Time', 1)
df = df.drop('First Transaction Time', 1)
df = df.drop('Correct Transaction Time', 1)
df = df.drop('Correct Step Duration (sec)', 1)
df = df.drop('Step Duration (sec)', 1)
df = df.drop('Error Step Duration (sec)', 1)
df = df.drop('Incorrects', 1)
df = df.drop('Corrects', 1)
df = df.drop('Hints', 1)
df = df.drop('KC(Default)', 1)
df = df.drop('Opportunity(Default)', 1)
```

```python
gp1 = df.groupby(['Anon Student Id','KC1'], as_index=False)['Opps1'].max()
    #this gets the max value of opp1 for each kc1 for each student
```

```python
gp1 = gp1.rename(index=str, columns={'KC1': 'KC', 'Opps1': 'Opp'})
```

```python
gp2 = df.groupby(['Anon Student Id','KC2'], as_index=False)['Opps2'].max()
    #this gets the max value of opp2 for each kc2 for each student
```

```python
gp2 = gp2.rename(index=str, columns={'KC2': 'KC', 'Opps2': 'Opp'})
```

```python
gp3 = pd.merge(gp1, gp2, on=['Anon Student Id', 'KC'], how='outer')
gp3['Opp'] = gp3[['Opp_x', 'Opp_y']].max(axis=1)
gp3['Opp'] = gp3['Opp'].astype(int)
gp3 = gp3.groupby(['Anon Student Id','KC'], as_index=False)['Opp'].max()
```

```python
gp = df.groupby(['Anon Student Id','KC7'], as_index=False)['Opps7'].max()
gp = gp.rename(index=str, columns={'KC7': 'KC', 'Opps7': 'Opp'})
```

```python
gp3 = pd.merge(gp, gp3, on=['Anon Student Id', 'KC'], how='outer')
gp3['Opp'] = gp3[['Opp_x', 'Opp_y']].max(axis=1)
gp3['Opp'] = gp3['Opp'].astype(int)
gp3 = gp3.groupby(['Anon Student Id','KC'], as_index=False)['Opp'].max()
```

```python
gp3
```

```python
hp1 = df.groupby(['Anon Student Id','KC1'], as_index=False)['Correct First Attempt'].sum()
hp2 = df.groupby(['Anon Student Id','KC2'], as_index=False)['Correct First Attempt'].sum()
hp1 = hp1.rename(index=str, columns={'KC1': 'KC', 'Correct First Attempt': 'CFA'})
hp2 = hp2.rename(index=str, columns={'KC2': 'KC', 'Correct First Attempt': 'CFA'})
```

```python
hp3 = pd.merge(hp1,hp2, on=['Anon Student Id', 'KC'], how='outer')
hp3['CFA'] = hp3[['CFA_x', 'CFA_y']].sum(axis=1)
hp3['CFA'] = hp3['CFA'].astype(int)
hp3 = hp3.groupby(['Anon Student Id','KC'], as_index=False)['CFA'].sum()
```

```python
hp = df.groupby(['Anon Student Id','KC7'], as_index=False)['Correct First Attempt'].sum()
hp = hp.rename(index=str, columns={'KC7': 'KC','Correct First Attempt': 'CFA'})
```

```python
hp3 = pd.merge(hp, hp3, on=['Anon Student Id', 'KC'], how='outer')
hp3['CFA'] = hp3[['CFA_x', 'CFA_y']].sum(axis=1)
hp3['CFA'] = hp3['CFA'].astype(int)
hp3 = hp3.groupby(['Anon Student Id','KC'], as_index=False)['CFA'].sum()
```

```python
hp3
```

```python
df_ratio = pd.merge(gp3, hp3, on=['Anon Student Id', 'KC'])
```

```python
df_ratio['Ratio'] = df_ratio['CFA']/df_ratio['Opp']
```

```python
df_ratio
```

```python
df_ratio['Passed'] = np.where(df_ratio['Ratio']>=.6, 1, 0)
```
