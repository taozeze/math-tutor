```python
import pandas as pd
import numpy as np
import csv as csv
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('algebra_2005_2006_train.txt', sep='\t')
```

```python
#add label columns for KC, Opp
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
KC_Labels = lb.fit_transform(df['KC(Default)'])
```

```python
s = pd.Series(KC_Labels, name="KC Labels")
```

```python
col = pd.DataFrame(s)
```

```python
df = pd.concat([df, col], axis=1)
```

```python
lb2 = LabelEncoder()
Opp_Labels = lb2.fit_transform(df['Opportunity(Default)'])
s = pd.Series(Opp_Labels, name="Opp Labels")
col = pd.DataFrame(s)
```

```python
df = pd.concat([df, col], axis=1)
df
```

```python
lb3 = LabelEncoder()
Step_Labels = lb3.fit_transform(df['Step Name'])
s = pd.Series(Step_Labels, name="Step Labels")
col = pd.DataFrame(s)
```

```python
df = pd.concat([df, col], axis=1)
```

```python
df_proc = df.drop('Opportunity(Default)', 1)
df_proc = df_proc.drop('KC(Default)', 1)
df_proc = df_proc.drop('Step Name', 1)
```

```python
df_proc = df_proc.drop('Step Start Time', 1)
df_proc = df_proc.drop('First Transaction Time', 1)
df_proc = df_proc.drop('Correct Transaction Time', 1)
df_proc = df_proc.drop('Step End Time', 1)
```

```python
df_proc = df_proc.drop('Anon Student Id', 1)
df_proc = df_proc.drop('Problem Hierarchy', 1)
df_proc = df_proc.drop('Problem Name', 1)
```

```python
df_proc = df_proc.drop('Correct Step Duration (sec)', 1)
df_proc = df_proc.drop('Error Step Duration (sec)', 1)
df_proc = df_proc.drop('Incorrects', 1)
df_proc = df_proc.drop('Corrects', 1)
df_proc = df_proc.drop('Hints', 1)
```

```python
df_proc = df_proc.fillna(0)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
n_samples = df_proc.shape[0]
df_train = df_proc[:n_samples / 2] # train on first half of dataset
df_test = df_proc[n_samples / 2:]    # test on last half of dataset
```

```python
target_col = 'Correct First Attempt'
feature_cols = list(df_proc.drop(target_col, axis=1)) 
```

```python
X_train = df_train[feature_cols]  # feature values 
y_train = df_train[target_col]

X_test = df_test[feature_cols]  # feature values 
y_test = df_test[target_col]  # corresponding targets/labels
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train) 
pred = clf.predict(X_test)
acc = accuracy_score(y_test, pred)
print acc
```
